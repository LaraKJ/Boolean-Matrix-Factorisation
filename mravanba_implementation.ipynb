```python
# Load ELSA data
elsa = pd.read_csv(r'C:\Users\s1989190\Documents\GitHub\Structure-of-Frailty-ELSA\Analysis\Data\elsa_wave9.csv')
deficits = elsa.loc[:, 'mob-walk100m': 'mem_day']
deficits = deficits.dropna() # Remove rows with missing values
```

```python
O = deficits.to_numpy() # Convert dataframe to a NumPy array
```

```python
# Initialise the MatrixCompletion class
K = 2  # Number of hidden dimensions
mask = None  # Provide a mask if needed
min_sum = True  # Adjust as needed
tol = 1e-4  # Tolerance for message updates
learning_rate = 0.1  # Damping parameter
max_iter = 1000  # Maximum number of message passing updates
verbose = True  # Set to True to see progress
p_x_1 = 0.3  # Prior probability of x=1
p_y_1 = 0.3  # Prior probability of y=1
p_1_given_1 = 0.99  # Probability of observing 1 for the input of 1
p_0_given_0 = 0.99  # Probability of observing 0 for the input of 0

bmf = MatrixCompletion(O, K, mask, min_sum, tol, learning_rate, max_iter, verbose, p_x_1, p_y_1, p_1_given_1, p_0_given_0)
bmf.run()
```

```python
# Access the factorized matrices
X = bmf.X  # Binary factorized matrix X
Y = bmf.Y  # Binary factorized matrix Y
Z = bmf.Z  # Reconstructed binary matrix Z
```
